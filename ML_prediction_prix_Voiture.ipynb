```python
data_df.columns
```

```python
#Colonne avec prix atuel
data_df=data_df[['Années', 'Prix_actuel', 'Kms_effectué', 'Propriétaire', 'Prix_vente',
       'Type_carburant_Diesel', 'Type_carburant_Petrol',
       'Type_vendeur_Individual', 'Transmission_Manual']]

#Colonne sans prix actuel
data_df=data_df[['Années', 'Kms_effectué', 'Propriétaire', 'Prix_vente',
       'Type_carburant_Diesel', 'Type_carburant_Petrol',
       'Type_vendeur_Individual', 'Transmission_Manual']]
```

```python
X= data_df.drop('Prix_vente', axis=1)
y=data_df.Prix_vente
```

```python
from sklearn.model_selection import train_test_split
X_train, xtest, ytrain, ytest = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(X_train, ytrain)
```

```python
predictions = model.predict(xtest)
```

```python
from sklearn import metrics
```

```python
print('RMSE', np.sqrt(metrics.mean_squared_error(ytest, predictions)))
```

```python
import pickle
file =open('random_forest_model.pkl', 'wb')
pickle.dump(model, file)
```

```python
#visualisation de notre modele de prediction
feat_importances = pd.Series(model.feature_importances_, index=X_train.columns)
feat_importances.nlargest(5).plot(kind='barh', figsize=(20, 5))
plt.show()
```
